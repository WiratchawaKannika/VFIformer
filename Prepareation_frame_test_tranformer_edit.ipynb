# Prepare text. file to predict frame

```python
import glob

for k in range(len(z_)) :
    #f_name_toSave = pth[:1]
    pth = f'{z_[k]}/'
    txt_name = pth.split('/')[7]
    f_name_toSave = pth.split('/')[:-2]
    f_name_toSave = ('/'.join(f_name_toSave))
    #path_text = glob.glob(f"{z_[k]}/*")
    #print(pth)
    # relative path to search all text files
    files = glob.glob(f"{pth}*")
    files.sort()
    df = pd.DataFrame(files, columns =['Path'])
    df_ = df[:-1]
    df_ = df_.reset_index(drop=True)
    df2 = pd.DataFrame(files, columns =['Path'])
    df2_ = df2[1:]
    df2_ = df2_.reset_index(drop=True)
    df_['Path_txt'] = ''
    for i in range(len(df_)):
        name1 = df_['Path'][i]
        name2 = df2_['Path'][i]
        df_.loc[df_.index[i], 'Path_txt'] = str(name1) + ' ' + str(name2) 
    list_path = df_['Path_txt'].tolist()
    with open(f'{f_name_toSave}/{txt_name}-demo.txt', 'w') as f:
         for line in list_path:
             f.write(f"{line}\n")
    print(f'On Process : Write text file name -> [ {f_name_toSave}/{txt_name}-demo.txt ] ')   
```
